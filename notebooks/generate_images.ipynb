### This notebook is for loading a finetuned model from memory and use it to synthesize images

```python
import torch
from diffusers import StableDiffusionPipeline
from safetensors.torch import load_file

# Load base model
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")

# Load checkpoint state
checkpoint = load_file("/usr/local/data/amarkr/lgcig/saved_models/finetuned/checkpoints/model.safetensors")

# Update UNet weights
pipe.unet.load_state_dict(checkpoint)
```

```python
# Generate image
prompt = "Chest X-ray of healthy patient"
image = pipe(prompt).images[0]
```
